**Missing values**

```python
print(chicago_df.isnull().sum())

```

```python
#Need to drop neighbourhood_group column from both df
chicago_df=chicago_df.drop('neighbourhood_group',axis=1)
New_df=New_df.drop('neighbourhood_group',axis=1)
```

```python
chicago_df.head()
```

```python
New_df.head()
```

```python
#Need to fiil na values in price column
chicago_df['price'].value_counts()


```

```python
median_price = chicago_df['price'].median()
chicago_df['price'].fillna(median_price, inplace=True)
```

```python
print("\nAfter filling missing values:")
print(chicago_df['price'].isnull().sum())
```

```python
#Need to change the last_review column datatype to datetime
chicago_df['last_review'] = pd.to_datetime(chicago_df['last_review'].astype(str), format="%Y-%m-%d")
```

```python
#Need to fill missing value of review_per_month column
chicago_df['reviews_per_month'].fillna(0, inplace=True)
```

```python
#Need to fill missing value of license column
chicago_df['license'].fillna('Not Provided', inplace=True)
```

```python
#Recheck after fiil missing values
print(chicago_df.isnull().sum())
```

```python
median_price = New_df['price'].median()
New_df['price'].fillna(median_price, inplace=True)
```

```python
#Need to change the last_review column datatype to datetime
New_df['last_review'] = pd.to_datetime(New_df['last_review'].astype(str), format="%Y-%m-%d")
```

```python
#Need to fill missing value of license column
New_df['license'].fillna('Not Provided', inplace=True)
```

```python
#Recheck after fiil missing values
print(New_df.isnull().sum())
```
